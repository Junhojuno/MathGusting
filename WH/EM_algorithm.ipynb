# EM ( Expectation Maximization )
 = 기댓값 최대화 알고리즘

## <사전 지식>
#### 1. 잠재변수, 내제변수 (Latent Variable) 
    : 구성개념이 직접적으로 관찰이나 측정되지 않는 변수

#### 2. 가우시안 혼합 모형
$p(x) = \sum_Z p(z)p(x|z) = \sum_{k=1}^{K} \theta_k \mathcal{N}(x|\mu_k, \Sigma_k)$
- $p(x)$ : 전체 분포 
- $p(x|z)$ : $z$(카테고리)분포일 경우에 $x$(가우시안)분포
- $p(z)$ : 카테고리 확률분포
- $\theta_k$ : 카테고리 확률 분포의 모수, 모든 성분 중 특정한 Z가 선택 될 확률

## <목적>
혼합모형에서 모수의 추정값을 찾는 반복적인 알고리즘
#### ex) 카테고리 X 가우시안 혼합모형

- 4가지 가우시안 모형


- 데이터 분포


## < EM >
$p(z|x)$ = responsibility = 각 데이터가 어떤 카테고리에 속하는가를 알려주는 조건부확률

> ##### ($\gamma_{ik}$ : $x_i$가 카테고리 $k$에서 만들어졌을 확률 )
$\begin{eqnarray}
\gamma_{ik} 
&=& p(z_i=k|x_i) \\
&=& \dfrac{p(z_i=k)p(x_i|z_i=k)}{p(x_i)} \\
&=& \dfrac{p(z_i=k)p(x_i|z_i=k)}{\sum_{k=1}^K p(x_i,z_i=k)} \\
&=& \dfrac{p(z_i=k)p(x_i|z_i=k)}{\sum_{k=1}^K p(z_i=k)p(x_i|z_i=k)} 
\end{eqnarray}$


가우시안혼합모형의 경우
> $\gamma_{ik} = \dfrac{\theta_k \mathcal{N}(x_i|\mu_k, \Sigma_k)}{\sum_{k=1}^K \theta_k \mathcal{N}(x_i|\mu_k, \Sigma_k)}$

위 식은 모수를 이용해 responsibility를 추정가능
> $(\theta_k, \mu_k, \Sigma_k) \;\; \implies \;\; \gamma_{ik}$

$\mu_k$로 미분
> $0 = - \sum_{i=1}^N \dfrac{p(z_i=k)p(x_i|z_i=k)}{\sum_{k=1}^K p(z_i=k)p(x_i|z_i=k)} \Sigma_k (x_i - \mu_k )$

>


## < EM step >
1. E step
    : 우리가 알고있는 모수가 정확하다고 가정하고 이를 사용하여 responsibility를 추정
2. M step
    : 우리가 알고있는 responsibility가 정확하다고 가정하고 이를 사용하여 모수값을 추정
