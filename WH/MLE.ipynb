# MLE ( Maximum Likelihood Estimation )
#### (최대우도추정)

## < 사전 지식 >

### 기본 기호
- $\theta$ = 모수의 집합 / $\mu$ = 기댓값 / $\sigma^2$ = 분산

### 모수 추정이란?
- 일반적으로 우리가 분석할 데이터는 표본이고 우리가 알고 싶은 것은 이 표본을 만들어낸 확률변수임
- 이 확률변수는 1)분포모형, 2)모수(기댓값, 분산 등) 를 가지고 있고 이것들을 알아내는 행위를 모수추청 또는 검정 이라고 함
---
- 1) 분포모형을 알아내는 행위 = 분포검정
    > 기본적으로 "확률변수의 분포가 정규분포다" 라는 것이 가설이고 이것을 검정하는 것이 정규성 검정
- 2) 모수를 알아내는 행위 = 모수검정과 모수추정
    > 기본적으로 "분포가 이미정해진 상태에서 "모수가 특정한 값을 가지는지 혹은 특정한 값과 비교해서 큰지 작은지"를 확인하는 과정이 모수검정
    
    > 모수가 실제로 어떤값을 가질 확률이 가장 높은 지를 알아내는 작업이 모수 추청

##### 여기서 우리가 공부할 부분은 모수 추정의 방법 중 MLE방법

## < 모수추정의 원리 > 

###  Likelihood 함수

$f(x;\theta)$ 
- $\theta$가 고정되 있을때 $x$값을 변수로 갖는 함수를 뜻함
- 일반적으로 우리는 데이터라는 표본$x$을 알고 모수$\theta$를 모름
- 그래서 우리는 $\mathcal{L}(\theta;x)$ 라는 함수를 만듦
    - $x$가 고정되어 있을때 $\theta$값을 변수로 하는 함수

###  MLE
주어진 샘플 $x$에 대해 likelihood를 가장 크게 해주는 모수 $\theta$를 찾는 방법

##### EX)
- [ 3, 3, 1, -1, 0, 0, 1, -2, 4, 1 ] 이라는 데이터가 존재할때 이 데이터가 아래 3가지 분포 중 어떤 분포에서 나왔을 확률이 가장높은가? 
(아래 예시는 각각 $\mu$가 -1, 0, 1 인 정규분포)

```python
plt.figure(figsize=(9,6))
x = np.linspace(-5, 5, 100)
plt.plot(x, sp.stats.norm(loc=0).pdf(x))
plt.scatter(1, 0, s=100, c='k')
plt.vlines(1, -0.05, 0.45, linestyle=":")
plt.text(1-0.09, -0.08, "$x_0$")
plt.plot(x, sp.stats.norm(loc=-1).pdf(x))
plt.plot(x, sp.stats.norm(loc=1).pdf(x))
plt.scatter(1, sp.stats.norm(loc=-1).pdf(1), s=100, c='g');
plt.scatter(1, sp.stats.norm(loc=0).pdf(1), s=100, c='b');
plt.scatter(1, sp.stats.norm(loc=1).pdf(1), s=100, c='r');
plt.text(-3.3, 0.35, "1. $f_X(x;-1)$")
plt.text(-0.5, 0.44, "2. $f_X(x;0)$")
plt.text(1.9, 0.35, "3. $f_X(x;1)$")
plt.show()
```

### MLE 구현
$\hat\theta_{\text{MLE}} = \arg \max_{\theta} L(\theta; \{x_i\})$
> 일반적으로 log를 붙여서 사용함 => 계산이 단순해짐

## < 정규분포의 모수추정 >

$p(x ; \theta ) = \mathcal{N}(x ; \mu, \sigma^2) = \dfrac{1}{\sqrt{2\pi\sigma^2}} \exp \left(-\dfrac{(x-\mu)^2}{2\sigma^2}\right)$

- N 번의 반복시행으로 각각의 데이터($=x_1, x_2,x_3, ...$)가 서로 독립적임
- 전체확률밀도함수는 각각의 확률밀도 함수의 곱과 같음

$L(\theta;x_1, \cdots, x_N) = p(x_1, \cdots, x_N;\theta) = \prod_{i=1}^N  \dfrac{1}{\sqrt{2\pi\sigma^2}} \exp \left(-\dfrac{(x_i-\mu)^2}{2\sigma^2}\right)$

- 미분을 쉽게하기 위해 로그를 취함

$\begin{eqnarray*}
\log L 
&=& \log p(x_1, \cdots, x_N;\theta)  
&=& -\dfrac{N}{2} \log(2\pi\sigma^2) - \dfrac{1}{2\sigma^2}\sum_{i=1}^N (x_i-\mu)^2
\end{eqnarray*}$

- 최댓값을 구하기위해 미분해서 0이 나오는 값을 찾음

$\begin{eqnarray*}
\dfrac{\partial \log L}{\partial \mu} 
&=& \dfrac{\partial}{\partial \mu} \left\{ \dfrac{N}{2} \log(2\pi\sigma^2) + \dfrac{1}{2\sigma^2}\sum_{i=1}^N (x_i-\mu)^2  \right\} = 0 \\
\dfrac{\partial \log L}{\partial \sigma^2} 
&=& \dfrac{\partial}{\partial \sigma^2} \left\{ \dfrac{N}{2} \log(2\pi\sigma^2) + \dfrac{1}{2\sigma^2}\sum_{i=1}^N (x_i-\mu)^2 \right\} = 0\\
\end{eqnarray*}$

- 기댓값
> $\mu = \dfrac{1}{N}\sum_{i=1}^N x_i = \bar{x}$

- 분산
> $\sigma^2  =  \dfrac{1}{N}\sum_{i=1}^N (x_i-\mu)^2 = \dfrac{1}{N}\sum_{i=1}^N (x_i-\bar{x})^2 = s^2$
