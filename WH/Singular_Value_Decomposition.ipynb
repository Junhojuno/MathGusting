# < SVD(Singular Value Decomposition) >
----

## SVD(=특잇값 분해)란...?

- A가 정방행렬이 아닐 경우 A를 3개의 행렬의 곱으로 나타내는 것
    - (어떤 행렬이 주어지더라도 가능)
- $A = U\Sigma V^T$, ( $A \in R^{N\times M}$ )
> *위 식은 아래와 같은 조건을 만족해야함
    - $Σ∈R^{N×M}$ : 대각성분이 양수인 대각행렬이어야 한다. 큰 수부터 작은 수의 순서로 배열한다.(대각행렬)
    - $U∈R^{N×N}$ :  모든 열벡터가 단위벡터이고 서로 직교해야 한다. (직교행렬)
    - $V∈R^{M×M}$:  모든 열벡터가 단위벡터이고 서로 직교해야 한다. (직교행렬)
        
        - $Σ$ 대각성분들 = 특이값
        - $U$ 열벡터들 = 왼쪽 특이벡터
        - $V$ 열벡터 = 오른쪽 특이벡터
- 특이값의 갯수는 행렬의 열과 행의 갯수 중 작은 값과 같다.


( $N > M$ )

$A = 
\begin{bmatrix}

\end{bmatrix}$

( $N < M$ )

$A = 
\begin{bmatrix}

# SVD의 기하학적의미
- 직교행렬=회전변환
- 대각행렬=스케일변환

    > 모든 행렬은 회전변환과 스케일 변환을 통해서 나타낼 수 있다.

#### 고유값 분해와의 비교
- 고유값 분해 : 고유벡터 있어서(=방향이 정해짐) 스케일만 변화함
- 특이값 분해 : 방향과 스케일이 변화함

## 특이값 분해의 축소형
지금까지한것은 full SVD라고 하고 일반적으로 reduced SVD를 함

- < full SVD >

    > 우리가 지금까지 한 것


- < thin SVD >

    > - $Σ$ 에서 대각파트가 아닌 0으로 구성된 부분을 없앰
    > - $U$ 에서 이에 대응되는 열벡터들을 제거


- < compact SVD >

    > 0인 singular value들까지 모두제거


- < truncated SVD >

    > - 0이 아닌 singular value까지 제거
    > - 따라서 원래의 A가 아닌 근사행렬 A'가 나옴
    > - $argmin ||A - A'||$ 을 통해 rank t 를 구해야함

#### 왜 하지?
- 데이터 압축, 노이즈제거 등에 활용
- PCA와 비슷
